```python
#OR bia _a_bit
# 3 = 11
# 2 = 10
#_ = 11
3 | 2
# pode perceber que os primeiros digitos são igual de ambas as variáveis
#quando comparado os segundo dígito, há uma diferenca.
#prevalece o valor 1 no segunto termo
# 1 ou 1 = 1 | 1 ou 0 = 1 - desse modo fica 1 e 1+ =11 =3
```

```python
#AND bia _a_bit
# 3 = 11
# 2 = 10
#_ = 10
3 & 2
# pode perceber que os primeiros digitos são igual de ambas as variáveis
#quando comparado os segundo dígito, há uma diferenca.
#prevalece o valor 1 no segunto termo
# 1 e 1 = 1 | 1 e 0 = 1 - desse modo fica 1 e 1+ =11 =3
```

```python
#XOR bia _a_bit
# 3 = 11
# 2 = 10
# _ = 01
3 ^ 2
# pode perceber que os primeiros digitos são igual de ambas as variáveis
#quando comparado os segundo dígito, há uma diferenca.
#prevalece o valor 1 no segunto termo
# 1 ou exclusivo 1 = 0 | 1 ou exclusivo 0 = 1 - desse modo fica 0 e 1+ =01 =1
```

```python
 from string import Template
print("Nome: %s Idade: %d" % (nome, idade))
print("Nome: %s Idade: %.3f" % (nome, idade))
print("Nome: %s Idade: %.3f %r %r" % (nome, idade, True, False))
print("Nome: %s Idade: %.3f %d %d" % (nome, idade, True, False))

print( "Nome :{0} Idade: {1}".format(nome, idade))
print( "Nome :{1} Idade: {0}".format(nome, idade))

print(f'Nome: {nome} Idade: {idade}')
print(f'Nome: {nome} Idade: {idade} {2**8+1}')

# importar template do string

s = Template("Nome: $nome Idade: $idade")
print(s.substitute(nome=nome, idade=idade))
#primeiro nome em (nome=nome) é o $nome(substituido $=s)
```
